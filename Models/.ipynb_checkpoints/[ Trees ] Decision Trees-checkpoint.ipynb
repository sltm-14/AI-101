# Decision Tree - Iris example

Asks a series of questions  
They perform a sequence of branching operations based on comparisons of some quantities  
They can perform aclassification and regression tasks  
They can perform multi output tasks  
They don't feature scaling or centering   

```python
import numpy as np 
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

np.random.seed(42)
```

## <font color=#50AAF5>Importing Data</font>

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier

iris = datasets.load_iris()
#print(iris.DESCR)
```

```python
print(type(iris))
```

### <font color=#6FC24D>Importing Data - Reshape</font>

```python
x = iris['data'][:,2:] # Petal lenght and width
y = iris['target']
```

## <font color=#50AAF5>Algorithm Implementation</font>

```python
tree_clf = DecisionTreeClassifier(max_depth = 2,random_state=42)
tree_clf.fit(x,y)
```

Visualize  the trained Decision Tree usin export_graphviz()

```python
from sklearn.tree import export_graphviz
dot_data = export_graphviz(tree_clf,
                          out_file=None,
                          feature_names=iris.feature_names[2:],
                          class_names=iris.target_names,
                          rounded=True,
                          filled=True)
```

```python
import pydotplus
from IPython.display import Image

graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png('iris_tree.png')
Image(filename='iris_tree.png')
```

## <font color=#50AAF5>Prediction</font>

Prediction for a %cm long and 1.5 cm wide petal

```python
tree_clf.predict_proba([[5,1.5]])
```

```python
iris.target_names
```
