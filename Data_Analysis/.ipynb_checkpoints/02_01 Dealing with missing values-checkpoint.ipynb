            Replace without changing the dataset

```python
df.dropna(subset = ["price"], axis = 0).head()
```

<font size="+1">
    <b>
        <p style="color:#4496C4;">
            Missing values
        </p>
    </b>
</font>

<b>
    <font size="+0">
        <p style="color:#90B736;">
            Replace "?" for NaN (N/A data)
        </p>
    </font>
</b>


```python
import numpy as np

df.replace("?", np.nan, inplace = True)
df.head(5)
```

<b>
    <font size="+0">
        <p style="color:#90B736;">
            notnull() method
        </p>
    </font>
</b>

```python
missing_data = df.notnull()
missing_data.head()
```

<b>
    <font size="+0">
        <p style="color:#90B736;">
            isnull() method
        </p>
    </font>
</b>

```python
missing_data = df.isnull()
missing_data.head()
```

<b>
    <font size="+0">
        <p style="color:#90B736;">
            Counting missing values
        </p>
    </font>
</b>

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

<b>
    <font size="+0">
        <p style="color:#90B736;">
            Replacing missing values with mean value
        </p>
    </font>
</b>

<b>
    <font size="+0">
        <p style="color:#BB8FCE;">
            Getting mean value
        </p>
    </font>
</b>

```python
df["normalized-losses"].head()
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean()
print("Average of normalized-losses:", avg_norm_loss)
```

<b>
    <font size="+0">
        <p style="color:#BB8FCE;">
            Replaceing "NaN" for mean value
        </p>
    </font>
</b>

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

<b>
    <font size="+0">
        <p style="color:#90B736;">
            Replacing missing values with the media
        </p>
    </font>
</b>

<b>
    <font size="+0">
        <p style="color:#BB8FCE;">
            Getting media value
        </p>
    </font>
</b>

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

<b>
    <font size="+0">
        <p style="color:#BB8FCE;">
            Replaceing "NaN" for media value
        </p>
    </font>
</b>

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

<b>
    <font size="+0">
        <p style="color:#EB7969;">
            Replacing missing all missing values
        </p>
    </font>
</b>

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)

df['num-of-doors'].value_counts()
```
