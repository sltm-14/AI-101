```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

# to make this notebook's output stable across runs, the seed will always get the same value
np.random.seed(42)

X = 2 * np.random.rand(100, 1)
#print(X)
y = 4 + 3 * X + np.random.randn(100, 1)
#print(y)
plt.scatter(X, y);
```

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression( fit_intercept=True, normalize=False, copy_X=True, n_jobs=None)
linear_reg.fit(X, y)
X_fit = np.linspace(0, 2, 100)
y_fit = linear_reg.predict(X_fit[:, np.newaxis])

plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14);
```

```python
linear_reg = LinearRegression( fit_intercept=False, copy_X=True, n_jobs=None)
linear_reg.fit(X, y)
X_fit = np.linspace(0, 2, 100)
y_fit = linear_reg.predict(X_fit[:, np.newaxis])

plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14);
```
