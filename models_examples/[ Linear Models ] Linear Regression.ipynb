# Linear Regresion

## <font color=#50AAF5>Data acquisition</font>

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
# to make this notebook's output stable across runs, the seed will always get the same value
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.scatter(X, y);
```

## <font color=#50AAF5>Model</font>

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression(fit_intercept=True)
linear_reg.fit(X, y)
```

## <font color=#50AAF5>Model plot</font>

```python
X_fit = np.linspace(0, 2, 100)
y_fit = linear_reg.predict(X_fit[:, np.newaxis])

plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14);
```

## <font color=#50AAF5>Prediction</font>

```python
X_new = np.array([[0], [2]])
X_new.shape
y_new = linear_reg.predict(X_new)
linear_reg.intercept_, linear_reg.coef_
```
