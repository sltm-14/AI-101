18. All the departments which have between 3 and 7 employees.

```python
%%sql
SELECT department_name, COUNT(*) as num_emp
FROM employees, departments
WHERE employees.department_id=departments.department_id
GROUP BY department_name
HAVING num_emp BETWEEN 3 AND 7;
```

20. City and number of employees of cities which have more employees than the average.

```python
%%sql
SELECT city, COUNT(employee_id) as num_emp
FROM employees, departments, locations
WHERE employees.department_id=departments.department_id
AND departments.location_id=locations.location_id
GROUP BY city HAVING num_emp>(
SELECT AVG(n_emp)
FROM (
SELECT COUNT(employees.employee_id) AS n_emp
FROM employees, departments, locations
WHERE employees.department_id=departments.department_id
AND departments.location_id=locations.location_id
GROUP BY city) AS n_emp);
```

```python
%%sql
SELECT AVG(n_emp)
FROM (
SELECT COUNT(employees.employee_id) AS n_emp
FROM employees, departments, locations
WHERE employees.department_id=departments.department_id
AND departments.location_id=locations.location_id
GROUP BY city) AS n_emp
```
